```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set_style("darkgrid")
```

```python
from sklearn.metrics import accuracy_score
```

```python
file_path = '/content/drive/MyDrive/ML/data/abalone/abalone.data'
df = pd.read_csv(file_path, header=None)
```

```python
#mục tiêu : phân loại các giới tính của các loài bào ngư thông qua chiều dài, cân nặng, tuổi (rings + 1.5)
```

```python
df.columns = [
    "Sex",
    "Length",
    "Diameter",
    "Height",
    "Whole weight",
    "Shucked weight",
    "Viscera weight",
    "Shell weight",
    "rings",
]
```

```python
df['ages'] = df.rings + 1.5
```

```python
df.head()
```

```python
num_rows, num_columns = df.shape
print(f"Số dòng: {num_rows}")
print(f"Số cột: {num_columns}")
```

```python
df['Sex'] = df['Sex'].map({'M': 0, 'F': 1, 'I': 2})
```

```python
df.head(10)
```

```python
X = df.iloc[:,1:10].values
y = df.iloc[:,0:1].values
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train, X_test, y_train, y_test = train_test_split(X,y,test_size=0.25,random_state=100)
```

##Gaussian Classifier

```python
from sklearn.naive_bayes import GaussianNB
```

```python
classifier = GaussianNB()
classifier.fit(X_train, y_train)
```

```python
y_pred = classifier.predict(X_test)
```

```python
accuracy = accuracy_score(y_test, y_pred)
print("Accuracy:", accuracy)
```

```python
# Hàm dự đoán giới tính từ một mẫu dữ liệu
def predict_gender_from_sample(sample, gender_classifier):
    # Chuyển mẫu cụ thể thành một dòng dữ liệu dạng DataFrame
    sample_df = pd.DataFrame([sample])
    predicted_gender = gender_classifier.predict(sample_df)

    # Ánh xạ nhãn dự đoán về giới tính ('M', 'F', 'I')
    gender_mapping = {0: 'M', 1: 'F', 2: 'I'}
    predicted_gender = [gender_mapping[gender] for gender in predicted_gender]

    return predicted_gender

# Mẫu cụ thể bạn muốn dự đoán giới tính
sample_to_test = {
    'Length': 0.420,
    'Diameter': 0.330,
    'Height': 0.110,
    'Whole weight': 0.600,
    'Shucked weight': 0.250,
    'Viscera weight': 0.150,
    'Shell weight': 0.180,
    'rings': 8,
    'ages': 9.5
}

predicted_gender = predict_gender_from_sample(sample_to_test, classifier)

print("Predicted Gender:", predicted_gender[0])
```
